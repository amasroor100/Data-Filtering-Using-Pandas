```python
hp2017.columns = hp2017.columns.str.replace("."," ").str.replace("\s+"," ").str.strip().str.upper()
```

```python
hp2016.columns = hp2016.columns.str.replace("."," ").str.replace("\s+"," ").str.replace("(","").str.replace(")","").str.replace("\s+"," ").str.strip().str.upper()
```

```python
hp2015.columns = hp2015.columns.str.replace("."," ").str.replace("\s+"," ").str.replace("(","").str.replace(")","").str.strip().str.upper()
```

```python
regions = pd.read_csv("region.csv")
```

```python
regions
```

```python
null_region = hp.loc[hp["REGION"].isnull(),"COUNTRY"]
```

```python
null_region
```

```python
for i in null_region:
    hp.loc[hp["COUNTRY"] == i, "REGION"] = [regions.loc[regions["COUNTRY"] == i, "REGION"]]
```

```python
hp
```

```python
hp.isnull().sum()
```

```python
dups = hp.duplicated(["COUNTRY","YEAR"])
```

```python
dups
```

```python
hp["COUNTRY"]=hp["COUNTRY"].str.upper()
```

```python
dups = hp.duplicated(["COUNTRY","YEAR"])
```

```python
hp[dups]
```

```python
hp.loc[hp["COUNTRY"]=="SOMALILAND REGION"]
```

```python
hp = hp.drop_duplicates(["COUNTRY","YEAR"])
```

```python
hp.notnull().sum().sort_values()
```

```python
hp = hp.dropna(thresh = 159,  axis = 1)
```

```python
hp
```

```python
sns.heatmap(hp.isnull(), cbar = False)

plt.show()
```
